# Module 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Loading the Dataset from the Excel file
file_path = "https://raw.githubusercontent.com/Teja220/EDA/main/EDA.xlsx"
df = pd.read_excel(file_path, sheet_name="EDA")

print("First 5 rows of the dataset:")
print(df.head())

# 2. Summary Statistics for Numerical and Categorical Data
print("\nSummary statistics for numerical data:")
print(df.describe())  # Summary for numerical features

# Checking for categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nCategorical Columns:")
print(categorical_cols)

# Summary for categorical columns
if len(categorical_cols) > 0:
    print("\nSummary for categorical columns:")
    print(df[categorical_cols].describe())

# 3. Data Cleaning (Handling Missing Values)
print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Dropping rows with missing data
df_cleaned = df.dropna()
print("\nData after dropping missing values:")
print(df_cleaned.head())

# Filling missing data using forward fill
df_filled = df.fillna(method='ffill')
print("\nData after forward fill:")
print(df_filled.head())

# 4. Visualization (for Numerical and Categorical Data)
# Count plot for GICS Sectors (Categorical data)
plt.figure(figsize=(12, 6))
sns.countplot(y=df['GICS Sector'], order=df['GICS Sector'].value_counts().index)
plt.title('Count of Companies in Each GICS Sector')
plt.xlabel('Count')
plt.ylabel('GICS Sector')
plt.show()

# Histogram for Founded Year
if 'Founded' in df.columns and pd.api.types.is_numeric_dtype(df['Founded']):
    plt.figure(figsize=(8, 6))
    df['Founded'].hist(bins=20)
    plt.title('Histogram of Founded Year')
    plt.xlabel('Year')
    plt.ylabel('Frequency')
    plt.show()

# 5. Measurement Scales (Nominal, Ordinal, Interval, Ratio)
# Nominal Scale Example: GICS Sector
print("\nValue counts for 'GICS Sector' column (Nominal scale example):")
print(df['GICS Sector'].value_counts())

# Ordinal Scale Example: Creating a ranking based on Founded Year
if 'Founded' in df.columns and pd.api.types.is_numeric_dtype(df['Founded']):
    df['Company Age Category'] = pd.qcut(df['Founded'], 5, labels=['Very Old', 'Old', 'Moderate', 'New', 'Very New'])
    print("\nFirst 5 rows with 'Company Age Category' (Ordinal scale example):")
    print(df[['Founded', 'Company Age Category']].head())

# 6. Comparing EDA with Classical and Bayesian Analysis
# Classical Analysis (T-test on Founded Year, assuming normality)
if 'Founded' in df.columns and pd.api.types.is_numeric_dtype(df['Founded']):
    t_stat, p_value = stats.ttest_1samp(df['Founded'].dropna(), popmean=1950)
    print(f"\nT-test on Founded Year: t-stat = {t_stat}, p-value = {p_value}")


# 7. List of Software Tools required for EDA
print("\nSoftware Tools for EDA:")
eda_software = [
    "Python (Programming Language)",
    "Jupyter Notebook / Jupyter Lab",
    "Pandas (Data Manipulation)",
    "NumPy (Numerical Computations)",
    "Matplotlib (Data Visualization)",
    "Seaborn (Statistical Data Visualization)",
    "Plotly (Interactive Visualizations)",
    "Scipy (Statistical Analysis)",
    "Scikit-learn (Machine Learning & Preprocessing)",
    "Statsmodels (Statistical Testing)",
    "Missingno (Missing Data Visualization)",
    "Sweetviz (Automated EDA)",
    "Pandas Profiling (EDA Automation)",
]

for software in eda_software:
    print(f"- {software}")
```

# Module 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# Load dataset
file_path = "https://raw.githubusercontent.com/Teja220/EDA/main/EDA.xlsx"
df = pd.read_excel(file_path, sheet_name="EDA")

print("First 5 rows of the dataset:")
print(df.head())

# Remove duplicates
print("\nShape before deduplication:", df.shape)
df.drop_duplicates(inplace=True)
print("Shape after deduplication:", df.shape)

# Convert 'Founded' column to numeric and handle missing values
df['Founded'] = pd.to_numeric(df['Founded'], errors='coerce')
missing_before = df['Founded'].isna().sum()
df['Founded'].fillna(df['Founded'].mean(), inplace=True)
missing_after = df['Founded'].isna().sum()

print(f"\nMissing values in 'Founded' column before imputation: {missing_before}")
print(f"Missing values after imputation: {missing_after}")
print(df[['Founded']].head())

# Bin 'Founded' into 5 categories
df['Founded_binned'] = pd.cut(df['Founded'], bins=5, labels=['Very Old', 'Old', 'Moderate', 'New', 'Very New'])

print("\nBinned 'Founded' column:")
print(df[['Founded', 'Founded_binned']].head())

# Introduce missing values for demonstration
df.loc[10:15, 'Founded'] = np.nan

# Impute missing values using mean (MLE approximation)
imputer = SimpleImputer(strategy='mean')
df['Founded'] = imputer.fit_transform(df[['Founded']])

print("\nFounded column after imputing missing values:")
print(df[['Founded']].head(20))

# Plot distribution of binned 'Founded' column
plt.figure(figsize=(8, 5))
df['Founded_binned'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Founded Year Bins')
plt.xlabel('Founded Year Bins')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Module 3

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
url = "https://raw.githubusercontent.com/Teja220/EDA/main/EDA.xlsx"
df = pd.read_excel(url, sheet_name="EDA")

# Convert 'Founded' to numeric
df['Founded'] = pd.to_numeric(df['Founded'], errors='coerce')

# 1. Univariate Analysis
print("Summary Statistics:\n", df.describe())

# Histogram for 'Founded' year
plt.figure(figsize=(10, 6))
df['Founded'].dropna().hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Company Founded Year')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 2. Bivariate Analysis (Correlation Matrix)
numerical_cols = df.select_dtypes(include=np.number).columns

if len(numerical_cols) > 1:
    print("\nCorrelation Matrix:\n", df[numerical_cols].corr())

    plt.figure(figsize=(12, 8))
    sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='rocket', linewidths=0.5, fmt=".2f", center=0)
    plt.title('Correlation Matrix Heatmap')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.show()

# Scatter Plot (Founded vs CIK)
if 'CIK' in df.columns and df['CIK'].dtype in [np.int64, np.float64]:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df['Founded'], y=df['CIK'], color='purple')
    plt.title('Founded Year vs CIK')
    plt.xlabel('Founded Year')
    plt.ylabel('CIK')
    plt.grid(True)
    plt.show()

# 3. Multivariate Analysis (Pairplot)
df_numeric = df[numerical_cols].dropna()  # Drop NaNs to avoid errors

sns.pairplot(df_numeric, diag_kind="hist", plot_kws={'s': 10, 'alpha': 0.5})
plt.suptitle("Multivariate Analysis (Pairplot)", fontsize=14, y=1.02)
plt.show()
```

# Time Series Analysis (TSA)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/Teja220/EDA/main/EDA.xlsx"
df = pd.read_excel(url, sheet_name="EDA")

# Convert 'Founded' to numeric and filter realistic years
df['Founded'] = pd.to_numeric(df['Founded'], errors='coerce')
df = df[df['Founded'].between(1700, 2025)].dropna(subset=['Founded'])
df['Founded'] = df['Founded'].astype(int)

# Count companies founded per year
yearly_counts = df['Founded'].value_counts().sort_index()

# Line Chart: Companies Founded Per Year
plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-', color='b', alpha=0.7)
plt.title('Number of Companies Founded Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Companies')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(yearly_counts.index.min(), yearly_counts.index.max(), 10), rotation=45)
plt.show()

# Moving Average (5-Year Rolling Mean)
rolling_window = 5
yearly_counts_smooth = yearly_counts.rolling(rolling_window).mean()

plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-', alpha=0.5, label='Actual Data')
plt.plot(yearly_counts_smooth.index, yearly_counts_smooth.values, linestyle='-', color='r', label=f'{rolling_window}-Year Moving Average')
plt.title(f'Companies Founded Over Time (Smoothed {rolling_window}-Year Avg)')
plt.xlabel('Year')
plt.ylabel('Number of Companies')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(yearly_counts.index.min(), yearly_counts.index.max(), 10), rotation=45)
plt.show()
```
